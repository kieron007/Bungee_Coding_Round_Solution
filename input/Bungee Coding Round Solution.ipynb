```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('main.csv')
df
```

### TASK 1
#### Create an output file "filteredCountry.csv" inside a folder named "output". This file should contain only those records where country contains the word USA

```python
CountryList = df.COUNTRY.unique()
CountryList
```

```python
USA_CountryList = []
for i in CountryList:
    if i[0]=='U' and i[1]=='S' and i[2]=='A':
        USA_CountryList.append(i)
print(USA_CountryList)
```

```python
filteredCountry = df[df.COUNTRY.isin(USA_CountryList)]
filteredCountry
```

```python
filteredCountry.to_csv('C:\\Users\\Kiran Shete\\Bungeetech\\intership-test-master\\output\\filteredCountry.csv', index = False)
```

### TASK 2
#### Now consider "filteredCountry.csv" as the input file. For each group of "SKU" find 2 minimum prices and store this result in "lowestPrice.csv" inside a folder named "output"

```python
SKU_List = filteredCountry.SKU.unique()
```

```python
filteredCountry.PRICE = filteredCountry.PRICE.str.replace('$','').str.replace('?','').str.replace(',','')
```

```python
filteredCountry['PRICE'] = filteredCountry['PRICE'].astype(float)
```

```python
SKU = []
FIRST_MINIMUM_PRICE = []
SECOND_MINIMUM_PRICE = []
for i in SKU_List:
    li = np.sort(filteredCountry[filteredCountry['SKU']==i].PRICE.unique())
    if len(li)>1:
        SKU.append(i)
        FIRST_MINIMUM_PRICE.append(li[0])
        SECOND_MINIMUM_PRICE.append(li[1])  
lowestPrice = pd.DataFrame({'SKU':SKU, 'FIRST_MINIMUM_PRICE':FIRST_MINIMUM_PRICE, 
                            'SECOND_MINIMUM_PRICE':SECOND_MINIMUM_PRICE}, )
lowestPrice
```

```python
lowestPrice.to_csv('C:\\Users\\Kiran Shete\\Bungeetech\\intership-test-master\\output\\lowestPrice.csv', index = False)
```
